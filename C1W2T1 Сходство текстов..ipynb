```python
f = open('sentences.txt', 'r')
```

```python
sentences = []

for line in f:
    sentences.append(line)

sentences
```

```python
f.close()
```

```python
low_sen = []

for line in sentences:
    low_sen.append(line.lower())
```

```python
import re

clean_lines = []

for line in low_sen:
    clean_lines.append(re.split('[^a-z]', line))
```

```python
lines = []

for line in clean_lines:
    lines.append(list(filter(lambda x: x is not '', line)))
    
lines
```

```python
lst = []

for line in lines:
    for l in line:
        lst.append(l)
        
lst_unique = []

for l in lst:
    if l not in lst_unique:
        lst_unique.append(l)

d = len(lst_unique)
n = len(lines)
```

```python
words = {i: lst_unique[i] for i in range(d)}
words
```

```python
import numpy as np
from collections import Counter
from collections import OrderedDict

my_dict = np.zeros((n, d))

for i in range(n):
    text_counts = Counter(lines[i])
    for j in range(d):
        '''string = ' ' + words[j] + ' '
        my_dict[i][j] = low_sen[i].count(string)
        del(string)'''
        my_dict[i][j] = text_counts[words[j]]
```

```python
results = np.zeros(n-1)
```

```python
import scipy.spatial as sp

for i in range(n-1):
    results[i] = sp.distance.cosine(my_dict[0], my_dict[i+1])
```

```python
first_sen_number = min(results)

for i in range(n-1):
    if first_sen_number == results[i]:
        ind_1 = i

results_2 = np.delete(results, ind_1)
second_sen_number = min(results_2)

for i in range(n-1):
    if second_sen_number == results[i]:
        ind_2 = i

ind_1 = ind_1 + 1
ind_2 = ind_2 + 1
```

```python
f = open('Task_1.txt', 'w')
f.write(str(ind_1) + ' ' + str(ind_2))
f.close()
```

```python
f = open('sentences.txt', 'r')

answers = []

for i, line in enumerate(f):
    if (i == ind_1 - 1) or (i == ind_2 - 1):
        answers.append(line)

f.close()
```
